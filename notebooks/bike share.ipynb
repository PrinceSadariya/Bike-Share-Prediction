### Now dataset is ready for model training

```python
from sklearn.model_selection import train_test_split
```

### model for prediction of casual

```python
Xdc = df_d.iloc[:,:8]
ydc = df_d.iloc[:,-2]
```

```python
Xdc
```

```python
ydc
```

```python
Xdc_train,Xdc_test,ydc_train,ydc_test = train_test_split(Xdc,ydc,test_size=0.3,random_state=43)
```

### I have checked with all models and found that the best model with the best parameter is GradientBoostingRegressor.

#### parameter is {'criterion': 'friedman_mse','learning_rate': 0.1,'loss': 'squared_error','max_features': 'sqrt','n_estimators': 128,'subsample': 0.6}

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr_dc = GradientBoostingRegressor(max_features='sqrt',n_estimators=128,subsample=0.6)
```

```python
gbr_dc.fit(Xdc_train,ydc_train)
```

```python
ydc_pred = gbr_dc.predict(Xdc_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(ydc_test,ydc_pred))
```
